```

### 4.5.1 Stima della media e della varianza da campioni

Assumiamo che i dati che osserviamo siano i.i.d. (indipendenti e identicamente distribuiti) e provenienti dalla stessa distribuzione sottostante ($ F $), con una media vera ($ \mu $) e una varianza vera ($ \sigma^2 $). Poiché non possiamo intervistare tutte le persone in Bhutan, dobbiamo basarci sul nostro campione per stimare media e varianza. Dal nostro campione possiamo calcolare una _media campionaria_ ($ \bar{X} $) e una _varianza campionaria_($ S^2 $). Queste rappresentano le migliori stime possibili della media vera e della varianza vera.

$$
\boxed{
    \begin{aligned}
    \text{Media Campionaria}\\
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i\\
    \end{aligned}
}
$$

$$
\boxed{
    \begin{aligned}
    \text{Varianza Campionaria}\\
S^2 = \frac{1}{n - 1} \sum_{i=1}^{n} (X_i - \bar{X})^2
\\
    \end{aligned}
}
$$

La prima domanda da porsi è: questi stimatori (media e varianza campionaia) sono corrette (non distorte)? Sì. Il termine "non distorta" significa che, se ripetessimo molte volte il processo di campionamento, il valore atteso delle nostre stime sarebbero uguali ai veri valori che stiamo cercando di stimare. 

Dimostriamo che ciò vale per $ \bar{X} $:

$$
\mathbb{E}[\bar{X}] = \mathbb{E}\left[\frac{1}{n} \sum_{i=1}^{n} X_i \right] = \frac{1}{n} \mathbb{E}\left[ \sum_{i=1}^{n} X_i \right] = \frac{1}{n} \sum_{i=1}^{n} \mathbb{E}[X_i] = \frac{1}{n} \sum_{i=1}^{n} \mu = \frac{1}{n} \cdot n\mu = \mu
$$

$\square$


Dimostriamo, per esercizio, che il valore atteso della varianza campionaria è uguale alla varianza reale:

Partiamo dalla definizione della varianza campionaria:
$$
S^2=\frac{1}{n-1}\sum_{i=1}^n (X_i-\bar X)^2.
$$

Calcoliamo prima l'identità algebrica che ci servirà. Sia $ \mu=\mathbb{E}[X_i] $. Si ha
$$
X_i-\bar X = (X_i-\mu)-(\bar X-\mu).
$$
Elevando al quadrato e sommando su (i):
$$
\sum_{i=1}^n (X_i-\bar X)^2
= \sum_{i=1}^n\big[(X_i-\mu)-(\bar X-\mu)\big]^2
= \sum_{i=1}^n (X_i-\mu)^2 -2(\bar X-\mu)\sum_{i=1}^n (X_i-\mu) + \sum_{i=1}^n (\bar X-\mu)^2.
$$
Ora osserviamo che $\sum_{i=1}^n (X_i-\mu)=n(\bar X-\mu)$ e che $\sum_{i=1}^n (\bar X-\mu)^2 = n(\bar X-\mu)^2$. Quindi
$$
\sum_{i=1}^n (X_i-\bar X)^2
= \sum_{i=1}^n (X_i-\mu)^2 -2n(\bar X-\mu)^2 + n(\bar X-\mu)^2
= \sum_{i=1}^n (X_i-\mu)^2 - n(\bar X-\mu)^2.
$$

Prendiamo ora il valore atteso su entrambi i membri. 
$$
\mathbb{E}\left[\sum_{i=1}^n (X_i-\bar X)^2\right]
= \mathbb{E}\left[\sum_{i=1}^n (X_i-\mu)^2\right] - n\,\mathbb{E}\left[(\bar X-\mu)^2\right].
$$
Usando linearità di $\mathbb{E}$ e il fatto che $\mathbb{E}[(X_i-\mu)^2]=\mathrm{Var}(X_i)=\sigma^2$ per ogni (i), otteniamo:
$$
\mathbb{E}\left[\sum_{i=1}^n (X_i-\mu)^2\right] = \sum_{i=1}^n \mathbb{E}[(X_i-\mu)^2] = n\sigma^2,
$$
e, dato che le $X_i$ sono i.i.d., abbiamo che  $\mathrm{Var}(\bar X)=\mathrm{Var}\left(\frac{1}{n}\sum_i X_i\right)=\frac{1}{n^2}\sum_i \mathrm{Var}(X_i)=\frac{\sigma^2}{n}$, quindi:
$$
\mathbb{E}\left[(\bar X-\mu)^2\right] = \mathrm{Var}(\bar X) = \frac{\sigma^2}{n},
$$

Sostituendo:
$$
\mathbb{E}\left[\sum_{i=1}^n (X_i-\bar X)^2\right] =   n\sigma^2 - n\cdot\frac{\sigma^2}{n} -  = (n-1)\sigma^2.
$$

Infine, dall'espressione di $S^2$:
$$
\mathbb{E}[S^2] = \mathbb{E}\left[\frac{1}{n-1}\sum_{i=1}^n (X_i-\bar X)^2\right]
= \frac{1}{n-1}\mathbb{E}\left[\sum_{i=1}^n (X_i-\bar X)^2\right]
= \frac{1}{n-1}\cdot (n-1)\sigma^2 = \sigma^2.
$$

Quindi $S^2$ è uno stimatore non distorto della varianza vera $ \sigma^2 $. $\square$

L'equazione della media campionaria sembra coerente con la nostra definizione di valore atteso. Lo stesso potrebbe dirsi per la varianza campionaria, a parte la presenza sorprendente di ($ n - 1 $) al denominatore. Perché proprio ($ n - 1 $)? Questo denominatore è necessario per garantire che $ \mathbb{E}[S^2] = \sigma^2 $.

L'intuizione alla base della dimostrazione è che la varianza campionaria calcola la distanza di ogni dato dalla media campionaria, non dalla media vera. La media campionaria stessa varia tra diversi campioni, e si può dimostrare che la sua varianza è legata alla varianza vera.
