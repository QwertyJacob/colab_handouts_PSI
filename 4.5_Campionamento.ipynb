### 4.5.1 Stima della media e della varianza da campioni

Assumiamo che i dati che osserviamo siano i.i.d. (indipendenti e identicamente distribuiti) e provenienti dalla stessa distribuzione sottostante ($ F $), con una media vera ($ \mu $) e una varianza vera ($ \sigma^2 $). Poiché non possiamo intervistare tutte le persone in Bhutan, dobbiamo basarci sul nostro campione per stimare media e varianza. Dal nostro campione possiamo calcolare una _media campionaria_ ($ \bar{X} $) e una _varianza campionaria_($ S^2 $). Queste rappresentano le migliori stime possibili della media vera e della varianza vera.

$$
\boxed{
    \begin{aligned}
    \text{Media Campionaria}\\
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i\\
    \end{aligned}
}
$$

$$
\boxed{
    \begin{aligned}
    \text{Varianza Campionaria}\\
S^2 = \frac{1}{n - 1} \sum_{i=1}^{n} (X_i - \bar{X})^2
\\
    \end{aligned}
}
$$

La prima domanda da porsi è: questi stimatori (media e varianza campionaia) sono corrette (non distorte)? Sì. Il termine "non distorta" significa che, se ripetessimo molte volte il processo di campionamento, il valore atteso delle nostre stime sarebbero uguali ai veri valori che stiamo cercando di stimare. 

Dimostriamo che ciò vale per $ \bar{X} $:

$$
\mathbb{E}[\bar{X}] = \mathbb{E}\left[\frac{1}{n} \sum_{i=1}^{n} X_i \right] = \frac{1}{n} \mathbb{E}\left[ \sum_{i=1}^{n} X_i \right] = \frac{1}{n} \sum_{i=1}^{n} \mathbb{E}[X_i] = \frac{1}{n} \sum_{i=1}^{n} \mu = \frac{1}{n} \cdot n\mu = \mu
$$

$\square$


Dimostriamo, per esercizio, che il valore atteso della varianza campionaria è uguale alla varianza reale:

Partiamo dalla definizione della varianza campionaria:
$$
S^2=\frac{1}{n-1}\sum_{i=1}^n (X_i-\bar X)^2.
$$

Calcoliamo prima l'identità algebrica che ci servirà. Sia $ \mu=\mathbb{E}[X_i] $. Si ha
$$
X_i-\bar X = (X_i-\mu)-(\bar X-\mu).
$$
Elevando al quadrato e sommando su (i):
$$
\sum_{i=1}^n (X_i-\bar X)^2
= \sum_{i=1}^n\big[(X_i-\mu)-(\bar X-\mu)\big]^2
= \sum_{i=1}^n (X_i-\mu)^2 -2(\bar X-\mu)\sum_{i=1}^n (X_i-\mu) + \sum_{i=1}^n (\bar X-\mu)^2.
$$
Ora osserviamo che $\sum_{i=1}^n (X_i-\mu)=n(\bar X-\mu)$ e che $\sum_{i=1}^n (\bar X-\mu)^2 = n(\bar X-\mu)^2$. Quindi
$$
\sum_{i=1}^n (X_i-\bar X)^2
= \sum_{i=1}^n (X_i-\mu)^2 -2n(\bar X-\mu)^2 + n(\bar X-\mu)^2
= \sum_{i=1}^n (X_i-\mu)^2 - n(\bar X-\mu)^2.
$$

Prendiamo ora il valore atteso su entrambi i membri. 
$$
\mathbb{E}\left[\sum_{i=1}^n (X_i-\bar X)^2\right]
= \mathbb{E}\left[\sum_{i=1}^n (X_i-\mu)^2\right] - n\,\mathbb{E}\left[(\bar X-\mu)^2\right].
$$
Usando linearità di $\mathbb{E}$ e il fatto che $\mathbb{E}[(X_i-\mu)^2]=\mathrm{Var}(X_i)=\sigma^2$ per ogni (i), otteniamo:
$$
\mathbb{E}\left[\sum_{i=1}^n (X_i-\mu)^2\right] = \sum_{i=1}^n \mathbb{E}[(X_i-\mu)^2] = n\sigma^2,
$$
e, dato che le $X_i$ sono i.i.d., abbiamo che  $\mathrm{Var}(\bar X)=\mathrm{Var}\left(\frac{1}{n}\sum_i X_i\right)=\frac{1}{n^2}\sum_i \mathrm{Var}(X_i)=\frac{\sigma^2}{n}$, quindi:
$$
\mathbb{E}\left[(\bar X-\mu)^2\right] = \mathrm{Var}(\bar X) = \frac{\sigma^2}{n},
$$

Sostituendo:
$$
\mathbb{E}\left[\sum_{i=1}^n (X_i-\bar X)^2\right] =   n\sigma^2 - n\cdot\frac{\sigma^2}{n} -  = (n-1)\sigma^2.
$$

Infine, dall'espressione di $S^2$:
$$
\mathbb{E}[S^2] = \mathbb{E}\left[\frac{1}{n-1}\sum_{i=1}^n (X_i-\bar X)^2\right]
= \frac{1}{n-1}\mathbb{E}\left[\sum_{i=1}^n (X_i-\bar X)^2\right]
= \frac{1}{n-1}\cdot (n-1)\sigma^2 = \sigma^2.
$$

Quindi $S^2$ è uno stimatore non distorto della varianza vera $ \sigma^2 $. $\square$

L'equazione della media campionaria sembra coerente con la nostra definizione di valore atteso. Lo stesso potrebbe dirsi per la varianza campionaria, a parte la presenza sorprendente di ($ n - 1 $) al denominatore. Perché proprio ($ n - 1 $)? Questo denominatore è necessario per garantire che $ \mathbb{E}[S^2] = \sigma^2 $.

L'intuizione alla base della dimostrazione è che la varianza campionaria calcola la distanza di ogni dato dalla media campionaria, non dalla media vera. La media campionaria stessa varia tra diversi campioni, e si può dimostrare che la sua varianza è legata alla varianza vera.

### 4.5.2 Errore Standard
Ora siamo convinti che le nostre stime di media e varianza non sono distorte. Ma ora vorremmo sapere di quanto la media campionaria potrebbe variare rispetto alla media vera.

$$
\text{Var}(\bar{X}) = \text{Var}\left( \frac{1}{n} \sum_{i=1}^{n} X_i \right) = \left( \frac{1}{n} \right)^2 \text{Var} \left( \sum_{i=1}^{n} X_i \right) = \left( \frac{1}{n} \right)^2 \sum_{i=1}^{n} \text{Var}(X_i) = \left( \frac{1}{n} \right)^2 \sum_{i=1}^{n} \sigma^2 = \left( \frac{1}{n} \right)^2 \cdot n\sigma^2 = \frac{\sigma^2}{n} \approx \frac{S^2}{n}
$$

$$
\text{Std}(\bar{X}) \approx \sqrt{ \frac{S^2}{n} }
$$

Il termine $ \text{Std}(\bar{X}) $ ha un nome speciale: si chiama **errore standard** ed è il modo in cui si riporta l'incertezza delle stime della media nei lavori scientifici (ed è ciò che genera le barre d'errore). 

Ottimo! Ora possiamo calcolare tutte queste statistiche interessanti per la popolazione del Bhutan. Ma un momento! Non abbiamo mai detto come calcolare l'errore standard della varianza campionaria, cioè, $ \text{Std}(S^2) $. Questo è più complicato perché il teorema del limite centrale non si applica al calcolo di $ S^2 $. Avremo invece bisogno di una tecnica più generale. 

Supponiamo di aver raccolto un campione di dimensione $ n = 200 $ persone. Supponiamo che la media campionaria risultasse $ \bar{X} = 83 $  e la varianza campionaria $ S^2 = 450 $. Possiamo ora calcolare l'errore standard della stima della media, ottenendo $ 1.5 $.

Quando riporteremo i risultati diremo che la nostra stima del punteggio medio di felicità in Bhutan è $ 83 \pm 1.5 $. La nostra stima della varianza della felicità è invece $ 450 \pm $?.

### 4.5.3 Bootstrap

Il *bootstrap* è una tecnica statistica recentemente sviluppata, utile sia per comprendere la distribuzione delle statistiche sia per calcolare i _p-value_ (un _p-value_ è la probabilità che un'affermazione scientifica sia errata). È stata inventata a Stanford nel 1979, in un periodo in cui i matematici stavano cominciando a capire come utilizzare i computer e le simulazioni al computer per approfondire la comprensione delle probabilità.

Il primo fondamentale spunto è il seguente: se avessimo accesso alla distribuzione sottostante $ F $, rispondere a quasi ogni domanda riguardo l'accuratezza delle nostre statistiche diverrebbe un compito semplice. Ad esempio, nella sezione precedente abbiamo fornito una formula per calcolare la varianza campionaria a partire da un campione di dimensione $ n $. Sappiamo che, in valore atteso, la nostra stima della varianza campionaria coincide con la varianza vera. Ma cosa succede se vogliamo conoscere la probabilità che la varianza vera sia entro un certo intervallo rispetto al valore calcolato? Questa domanda può sembrare tecnica, ma è fondamentale per valutare affermazioni scientifiche! Se conoscessimo la distribuzione sottostante $ F $, potremmo semplicemente ripetere l'esperimento: estrarre un campione di dimensione $ n $ da $ F $, calcolare la varianza campionaria su questo nuovo campione e verificare in quale proporzione dei casi il risultato cade entro un certo intervallo.

Il secondo fondamentale spunto del *bootstrap* è che la migliore stima possibile di $ F $ la otteniamo direttamente dal nostro campione! Il modo più semplice per stimare $ F $ (e quello che utilizzeremo in questo corso) è assumere che $ P(X = k) $ sia semplicemente la frazione di volte in cui il valore $ k $ è apparso nel campione. Notiamo che ciò definisce la funzione di massa di probabilità della nostra stima $ \hat{F} $ di $ F $.

````python
import numpy as np

def bootstrap(): 
    N = 200 # numero di elementi nel campione 
    vals = np.arange(1,101, dtype=int)
    pmf = np.array([1/100] * 100) # stima della funzione di massa di probabilità sottostante a partire dal campione 
    stats = [] 
    for i in range(10000): 
        resample = np.choice(vals, size=N, p=pmf) # estrai N nuovi campioni dalla pmf 
        stat = np.var(resample) # calcola la statistica di interesse sul resample )in questo caso la varianza campionaria
        stats.append(stat) # stats può ora essere usato per stimare la distribuzione della statistica
````

Il metodo di *bootstrap* è ragionevole perché il campione in nostro possesso rappresenta la migliore — e unica — informazione disponibile sulla forma reale della distribuzione della popolazione. Inoltre, la maggior parte dei campioni, a patto che siano scelti casualmente, tende a somigliare abbastanza alla popolazione da cui provengono.

Per calcolare $ \text{Var}(S^2) $, potremmo calcolare $ S^2_i $ per ogni resample $ i $ e, dopo 10.000 iterazioni, calcolare la varianza campionaria di tutti i valori $ S^2_i $. Potreste chiedervi perché il resample deve avere la stessa dimensione del campione originale ($ n $). La risposta è che la variabilità della statistica che state calcolando può dipendere dalla dimensione del campione (o del resample). Per stimare in modo accurato la distribuzione della statistica, dobbiamo utilizzare resample della stessa dimensione.

Il *bootstrap* gode di solide garanzie teoriche ed è ampiamente accettato dalla comunità scientifica. Tuttavia, perde di validità quando la distribuzione sottostante ha una "coda lunga" (long tail) oppure quando i campioni non sono indipendenti e identicamente distribuiti (i.i.d.).

````python
p = 0.77  # Proporzione Campionaria
sample_size = 34  # Size del campione
num_extractions = 10000  # Numero di campionamenti simulati ;)

sample_variances = []
sample_means = []

def bootstrap():

    # Per ogni campionamento:
    for _ in range(num_extractions):
        # Simulazione del campionamento:
        X = np.random.binomial(1, p, sample_size)
        # Calcolo e salvo la media campionaria
        sample_means.append(np.mean(X))
        # Calcolo e salvo la varianza campionaria
        sample_variance = np.var(X, ddof=1)  # ddof=1 -> dividere per(n-1)
        sample_variances.append(sample_variance)


bootstrap()
````

````python
"""
Dato un campione (stavolta di dimensione 10000)
di varianze campionarie, mi calcolo a sua volta
la varianza campionaria di quel campione
"""
meta_variance = np.var(sample_variances, ddof=1)

"""
La radice quadrata di tale meta-varianza campionaria
sarà la meta-deviazione standard

Cioè l'errore standard della varianza campionaria.
"""
std_err_sample_var = np.sqrt(meta_variance)

print(f'Stima per la varianza: {np.mean(sample_variances):.5f}')
print(f'Errore standard della stima per la varianza: {std_err_sample_var:.5f}')
````
