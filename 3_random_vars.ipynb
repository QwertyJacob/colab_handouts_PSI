> **DEFINIZIONE 3.1** (Variabile Aleatoria Discreta)
Una variabile aleatoria è una funzione di uno **esito**,  
$ X = f ( \omega ) $.  
 
Il dominio di una variabile aleatoria è lo **spazio campione** $ \Omega $. Il suo codominio può essere l'insieme di tutti i numeri reali $ \mathbb{R} $, oppure solo i numeri positivi $ (0, +\infty) $, oppure gli interi $ \mathbb{Z} $, oppure l'intervallo $ (0, 1) $, ecc., a seconda dei valori possibili che la variabile aleatoria può assumere.

Una volta completato un esperimento e noto l'esito $ \omega $, il valore della variabile aleatoria $ X(\omega) $ risulta determinato.

In altre parole, dato uno spazio campione $\Omega$, e un insieme numerabile di numeri reali $D$, una variabile aleatoria discreta è una funzione con dominio $\Omega$ e codominio $D$. (Essa associa ad ogni elemento di $\Omega$ un numero in $D$.

Questo significa che per ogni risultato $\omega$ c'è un numero associato ad esso tramite la variabile aleatoria $X$, che potremmo scrivere come $X(\omega)$. Nota: non stiamo parlando della probabilità dell'esito, ma di un'altro numero, qualsiasi, che viene associato all'occorrenza dell'esito in questione.

Ora parliamo di **probabilità**: La probabilità che una variabile aleatoria $X$ assuma il valore $x$ è data dalla probabilità che si verifichi l'evento $\{\omega : X(\omega) = x\}$, Cioè $P(\{\omega : X(\omega) = x\})$. Questa probabilità è più comunemente scritta $P(\{X = x\})$, ma ancor più comunemente scritta così:  $P(X = x)$, e, moltissime volte, per brevità, anche così: $P(x)$.

> Quindi non farti fregare! il fatto che una v.a. assuma un certo valore è un evento, oppure rappresenta un evento, che ha una certa probabilità. Se poi questo evento fa riferimento ad un singolo esito oppure ad un insieme di molti esiti, dipenderà di come abbiamo modellato la v.a. in questione.

**Distribuzione di Probabilità di una Variabile Aleatoria Discreta**
La distribuzione di probabilità di una variabile aleatoria discreta è l'insieme dei numeri $P(\{X = x\})$ per ogni valore $x$ che $X$ può assumere. La distribuzione assume il valore 0 per tutti gli altri numeri. Si noti che la distribuzione è non negativa. La distribuzione di probabilità è talvolta nota come funzione di massa di probabilità. Anche abbreviata con la sigla PMF (dall'inglese _probability mass function_).

> **DEFINIZIONE 3.2**  (**Funzione di Massa di Probabilità**)  La funzione  $ P (x) = P ( X = x ) $  è la **funzione di massa di probabilità**, o **PMF**.  

> **DEFINIZIONE 3.2BIS** (Funzione di distribuzione Cumulativa) La **funzione di distribuzione cumulativa**, o **cdf**, è definita come:

$$
F (x) = P ( X \leq x ) = \sum_{y \leq x} P (y).
\quad (3.1)
$$

L'insieme dei valori possibili di $ X $ è detto **supporto** della distribuzione $ F $.


**Esempio(BONUS)** (Numeri da Monete III) Lanciamo una moneta truccata 2 volte. I lanci sono indipendenti. La moneta ha $P(T) = p$, $P(C) = 1-p$. Questo produce un numero aleatoria a 2 bit, che, convertita a base decimale, è una variabile aleatoria che assume i valori 0, 1, 2, 3. Qual è la distribuzione di probabilità di questa variabile aleatoria?

*Soluzione* Distribuzione di probabilità:
- $P(0) = (1-p)^2$
- $P(1) = (1-p)p$
- $P(2) = p(1-p)$
- $P(3) = p^2$



**Esempio(BONUS)** (Scommettere sulle Monete) Un modo per ottenere una variabile aleatoria è pensare alla ricompensa di una scommessa. Concordiamo di giocare il seguente gioco. Lancio una moneta. La moneta ha $P(T) = p$, $P(C) = 1-p$. Se la moneta esce testa, mi paghi $q$; se la moneta esce croce, ti pago $r$. Il numero di dollari che cambiano di mano è una variabile aleatoria. Qual è la sua distribuzione di probabilità?

*Soluzione* Vediamo questo problema dalla mia prospettiva. Se la moneta esce testa, ottengo $q$; se esce croce, ottengo $-r$. Quindi abbiamo $P(X = q) = p$ e $P(X = -r) = (1-p)$, e tutte le altre probabilità sono zero.


Per ogni esito $ \omega $, la variabile $ X $ assume uno e un solo valore $ x $.  
Questo implica che gli eventi $ \{ X = x \} $ sono disgiunti e esaustivi, e pertanto  

$$
\sum_x P (x) = \sum_x P ( X = x ) = 1.
$$



Osservando (3.1), possiamo concludere che la funzione $ F (x) $ è non decrescente in $ x $, assume sempre valori compresi tra 0 e 1, e soddisfa  

$$
\lim_{x \to -\infty} F (x) = 0 \quad \text{e} \quad \lim_{x \to +\infty} F (x) = 1.
$$

Tra due valori successivi di $ X $, la funzione $ F (x) $ è costante.  

In corrispondenza di ogni valore possibile $ x $ di $ X $, la funzione presenta uno "spike" pari a $ P (x) $ (vedi Fig. 3.1, a destra).

Ricordiamo che un modo per calcolare la probabilità di un evento è sommare le probabilità di tutti gli esiti che lo compongono.  
Di conseguenza, per ogni insieme $ A $,  

$$
P ( X \in A ) = \sum_{x \in A} P (x).
$$

Quando $ A $ è un intervallo, la sua probabilità può essere calcolata direttamente dalla funzione $ F (x) $:  

$$
P ( a < X \leq b ) = F (b) - F (a).
$$

**Esempio 3.3 (Errori in moduli indipendenti).**  
Un programma è composto da due moduli.  
Il numero di errori $ X_1 $ nel primo modulo ha funzione di massa $ P_1(x) $, e il numero di errori $ X_2 $ nel secondo modulo ha funzione di massa $ P_2(x) $, indipendentemente da $ X_1 $, dove:

| $ x $ | $ P_1(x) $ | $ P_2(x) $ |
|--------|-------------|-------------|
| 0      | 0.5         | 0.7         |
| 1      | 0.3         | 0.2         |
| 2      | 0.1         | 0.1         |
| 3      | 0.1         | 0.0         |

Si determini la funzione di massa di probabilità (pmf) e la funzione di distribuzione cumulativa (cdf) di $ Y = X_1 + X_2 $, il numero totale di errori.

**Soluzione.**  
Suddividiamo il problema in passi.  
Prima determiniamo tutti i valori possibili di $ Y $, quindi calcoliamo la probabilità di ciascun valore.

Chiaramente, $ Y $ è un intero compreso tra $ 0 + 0 = 0 $ e $ 3 + 2 = 5 $.  
Poiché $ P_2(3) = 0 $, il secondo modulo ha al massimo 2 errori.

Calcoliamo le probabilità:

$$
P_Y(0) = P ( Y = 0 ) = P ( X_1 = 0, X_2 = 0 ) = P_1(0) P_2(0) = (0.5)(0.7) = 0.35
$$

$$
P_Y(1) = P ( Y = 1 ) = P_1(0)P_2(1) + P_1(1)P_2(0) = (0.5)(0.2) + (0.3)(0.7) = 0.10 + 0.21 = 0.31
$$

$$
P_Y(2) = P ( Y = 2 ) = P_1(0)P_2(2) + P_1(1)P_2(1) + P_1(2)P_2(0) = (0.5)(0.1) + (0.3)(0.2) + (0.1)(0.7) = 0.05 + 0.06 + 0.07 = 0.18
$$

$$
P_Y(3) = P ( Y = 3 ) = P_1(1)P_2(2) + P_1(2)P_2(1) + P_1(3)P_2(0) = (0.3)(0.1) + (0.1)(0.2) + (0.1)(0.7) = 0.03 + 0.02 + 0.07 = 0.12
$$

$$
P_Y(4) = P ( Y = 4 ) = P_1(2)P_2(2) + P_1(3)P_2(1) = (0.1)(0.1) + (0.1)(0.2) = 0.01 + 0.02 = 0.03
$$

$$
P_Y(5) = P ( Y = 5 ) = P_1(3)P_2(2) = (0.1)(0.1) = 0.01
$$

Verifichiamo ora che la somma delle probabilità sia 1:

$$
\sum_{y=0}^{5} P_Y(y) = 0.35 + 0.31 + 0.18 + 0.12 + 0.03 + 0.01 = 1.00
$$

Pertanto, siamo probabilmente riusciti a considerare tutti i casi possibili senza omissioni.  
(Vogliamo sottolineare che il semplice fatto di ottenere $ \sum P(x) = 1 $ non garantisce l'assenza di errori nella soluzione. Tuttavia, se questa uguaglianza non fosse soddisfatta, allora avremmo certamente commesso un errore.)

La funzione di distribuzione cumulativa $ F_Y(y) $ può essere calcolata come segue:

$$
F_Y(0) = P_Y(0) = 0.35
$$

$$
F_Y(1) = F_Y(0) + P_Y(1) = 0.35 + 0.31 = 0.66
$$

$$
F_Y(2) = F_Y(1) + P_Y(2) = 0.66 + 0.18 = 0.84
$$

$$
F_Y(3) = F_Y(2) + P_Y(3) = 0.84 + 0.12 = 0.96
$$

$$
F_Y(4) = F_Y(3) + P_Y(4) = 0.96 + 0.03 = 0.99
$$

$$
F_Y(5) = F_Y(4) + P_Y(5) = 0.99 + 0.01 = 1.00
$$

Tra i valori di $ Y $, la funzione $ F_Y(x) $ è costante.

## Il Valore Atteso e le Sue Proprietà Fondamentali

La distribuzione di una variabile aleatoria, ossia l'intera collezione delle probabilità ad essa associata, contiene tutte le informazioni sul suo comportamento.  
Queste informazioni dettagliate possono essere sintetizzate in alcune caratteristiche fondamentali, come il valore medio o _valore atteso_, il valore più probabile della variabile aleatoria, la sua dispersione, variabilità, ecc.  

Le misure più comunemente utilizzate sono il valore atteso, la varianza, la deviazione standard.

> **DEFINIZIONE 3.5** **(Valore atteso o media per variabili aleatorie discrete)**  Data una variabile aleatoria discreta $X$ che assume valori in un insieme numerabile $D$ con funzione di massa di probabilità $P(X = x)$, il valore atteso è definito come:

$$\mathbb{E}[X] = \mu = \sum_{x \in D} x P(X = x)$$

dove la somma è estesa a tutti i possibili valori che la variabile aleatoria può assumere. Il valore atteso è spesso indicato con la lettera greca $ \mu $.  

In un certo senso, il valore atteso rappresenta la migliore previsione di $ X $.  La variabile stessa è aleatoria: assume diversi valori con diverse probabilità $ P(x) $.  
Allo stesso tempo, possiede un solo valore atteso $ E(X) $, che è non aleatorio.

**Osservazioni Importanti:**
1. Il valore atteso potrebbe non esistere se la serie non converge assolutamente
2. Il valore atteso può non appartenere all'insieme dei valori possibili della variabile aleatoria
3. $\mathbb{E}[X]$ rappresenta il "centro di massa" della distribuzione di probabilità
4. Sappiamo che $ X $ può assumere diversi valori con diverse probabilità. Per questo motivo, il suo valore atteso non è semplicemente la media aritmetica dei suoi valori, ma piuttosto una **media ponderata**.

**Esempio 3.7.** Si consideri una variabile che assume i valori 0 e 1 con probabilità $ P(0) = P(1) = 0.5 $. Cioè,

$$
X = 
\begin{cases}
0 & \text{con probabilità } 1/2 \\
1 & \text{con probabilità } 1/2
\end{cases}
$$

Osservando questa variabile molte volte, vedremo $ X = 0 $ circa il 50% delle volte e $ X = 1 $ circa il 50% delle volte.  
Il valore atteso di $ X $ sarà quindi vicino a 0.5, per cui è ragionevole definire $ E(X) = 0.5 $.

♦

Vediamo ora come probabilità diverse per $ X = 0 $ e $ X = 1 $ influenzino il valore atteso $ E(X) $.

**Esempio 3.8.** Supponiamo che $ P(0) = 0.75 $ e $ P(1) = 0.25 $.  
Allora, a lungo andare, $ X $ assume valore 1 solo 1 volta su 4, e valore 0 negli altri casi.  
Supponiamo di guadagnare 1 dollaro ogni volta che osserviamo $ X = 1 $.  
In media, guadagniamo 1 dollaro ogni quattro osservazioni, cioè 0.25 dollari per osservazione.  
Pertanto, in questo caso, $ E(X) = 0.25 $.

### Valore atteso di una funzione

Spesso siamo interessati a un'altra variabile $ Y $, che è funzione di $ X $.  
Ad esempio, il tempo di download dipende dalla velocità di connessione, il profitto di un negozio di computer dipende dal numero di computer venduti, e il bonus del suo gestore dipende da tale profitto.

Il valore atteso di $ Y = g(X) $ si calcola mediante una formula analoga:

$$
E ( g(X) ) = \sum_x g(x) P (x).
\quad (3.4)
$$

**Osservazione:**  
Se $ g $ è una funzione biunivoca (iniettiva), allora $ Y $ assume ciascun valore $ y = g(x) $ con probabilità $ P(x) $, e la formula per $ E(Y) $ può essere applicata direttamente.  Se invece $ g $ non è biunivoca, alcuni valori di $ g(x) $ si ripeteranno nella somma (3.4).  
Tuttavia, ciascuna occorrenza è moltiplicata per la corrispondente probabilità $ P(x) $.  
Quando si effettua la somma in (3.4), le probabilità associate allo stesso valore $ y = g(x) $ vengono sommate implicitamente, per cui ogni valore di $ g(x) $ risulta comunque moltiplicato per la probabilità $ P_Y(y) $, dove $ P_Y $ è la funzione di massa di $ Y $.  
Pertanto, la formula (3.4) è valida in generale, indipendentemente dal fatto che $ g $ sia iniettiva o meno.

Questa regola è anche detta legge dello statistico inconscio, o LOTUS, ci permette di calcolare il valore atteso di una funzione di una variabile aleatoria senza dover prima trovare la distribuzione della funzione stessa.

**Importanza della LOTUS**
Questa legge è particolarmente utile quando:
1. Vogliamo calcolare il valore atteso di una trasformazione di $X$
2. Non è pratico o possibile trovare la distribuzione di $g(X)$
3. Abbiamo già la distribuzione di $X$

> **Esempio(BONUS)** Se volessimo calcolare $\mathbb{E}[X^2]$ per una variabile aleatoria discreta $X$:

$$\mathbb{E}[X^2] = \sum_{x \in D} x^2 P(X = x)$$

anziché dover prima trovare la distribuzione di $X^2$.

## Proprietà del Valore Atteso:

Le seguenti proprietà lineari dei valori attesi seguono direttamente dalle formule (3.3) e (3.4).  
Per ogni coppia di variabili aleatorie $ X $ e $ Y $, e per ogni numeri non aleatori $ a $, $ b $ e $ c $, vale:

$$
E ( aX + bY + c ) = a E(X) + b E(Y) + c
\quad (3.5)
$$

In particolare:

$$
E ( X + Y ) = E(X) + E(Y)
$$

$$
E ( aX ) = a E(X)
$$

$$
E ( c ) = c
$$

**Caso Generale del valore atteso dellaa somma di variabili aleatorie:**
Per una somma finita di variabili aleatorie $X_1, X_2, \ldots, X_n$:

$$\mathbb{E}\left[\sum_{i=1}^n X_i\right] = \sum_{i=1}^n \mathbb{E}[X_i]$$

**Valore atteso della multiplicazione di variabili aleatorie indipendenti**:
Nel caso di variabili aleatorie $ X $ e $ Y $ indipendenti, vale inoltre:

$$
E ( XY ) = E(X) E(Y)
$$
**ATTENZIONE:** L'ultima proprietà derivata dalla (3.5) può valere anche per alcune variabili dipendenti, pertanto **non può essere usata** per verificare l'indipendenza tra $ X $ e $ Y $.

## Esempi sul Valore Atteso

Parliamo ancora dell'esempio precedente. La moneta ha $P(T) = p$, $P(C) = 1-p$. Se la moneta esce testa, mi paghi $q$; se la moneta esce croce, ti pago $r$. Ora immagina che giochiamo questo gioco molte volte. La nostra definizione frequentista della probabilità implica che in $N$ giochi, ci aspettiamo di vedere circa $pN$ teste e $(1-p)N$ croci. A sua volta, questo significa che il mio reddito totale da questi $N$ giochi dovrebbe essere circa $(pN)q - ((1-p)N)r$. La $N$ in questa espressione è scomoda; invece, potremmo dire che per ogni singolo gioco, il mio reddito atteso è:

$pq - (1-p)r$

Questo non è il reddito effettivo da un singolo gioco (che sarebbe o $q$ o $r$, a seconda di cosa ha fatto la moneta). È invece una stima di quello che accadrebbe su un gran numero di giochi, su base _per-gioco_. Questo è un esempio di un valore atteso.

> Si noti che il valore atteso potrebbe assumere un valore che la variabile aleatoria non assume. **Esempio** (Scommettere sulle Monete) Concordiamo di giocare il seguente gioco. Lancio una moneta equa (cioè $P(T) = P(C) = 1/2$). Se la moneta esce testa, mi paghi 1; se la moneta esce croce, ti pago 1. # La Varianza di una Variabile Aleatoria

## Definizione

La varianza di una variabile aleatoria $X$ è una misura della sua dispersione attorno al valore atteso $\mathbb{E}[X]$. È definita come:

$$\text{Var}[X] = \mathbb{E}[(X - \mathbb{E}[X])^2]$$

Intuitivamente, misura la media dei quadrati delle deviazioni dal valore atteso.

## Proprietà Fondamentali

1. **Non-negatività**:
   $$\text{Var}[X] \geq 0$$

2. **Varianza di una costante**:
   $$\text{Var}[c] = 0$$
   dove $c$ è una costante

3. **Effetto di una costante moltiplicativa**:
   $$\text{Var}[aX] = a^2\text{Var}[X]$$
   dove $a$ è una costante

4. **Effetto di una traslazione**:
   $$\text{Var}[X + b] = \text{Var}[X]$$
   dove $b$ è una costante

5. **Varianza della somma di variabili aleatorie indipendenti**:
   $$\text{Var}[X + Y] = \text{Var}[X] + \text{Var}[Y]$$
   (questa proprietà vale solo se $X$ e $Y$ sono indipendenti)

## Dimostrazione della Formula Alternativa

Dimostriamo ora che $\text{Var}[X] = \mathbb{E}[X^2] - \mathbb{E}[X]^2$

* Passo 1: Partiamo dalla definizione
$$\text{Var}[X] = \mathbb{E}[(X - \mathbb{E}[X])^2]$$

* Passo 2: Espandiamo il quadrato
$$(X - \mathbb{E}[X])^2 = X^2 - 2X\mathbb{E}[X] + \mathbb{E}[X]^2$$

* Passo 3: Applichiamo il valore atteso
$$\text{Var}[X] = \mathbb{E}[X^2 - 2X\mathbb{E}[X] + \mathbb{E}[X]^2]$$

* Passo 4: Usiamo la linearità del valore atteso
$$\text{Var}[X] = \mathbb{E}[X^2] - 2\mathbb{E}[X]\mathbb{E}[X] + \mathbb{E}[X]^2$$

* Passo 5: Semplifichiamo
$$\text{Var}[X] = \mathbb{E}[X^2] - 2(\mathbb{E}[X])^2 + (\mathbb{E}[X])^2 = \mathbb{E}[X^2] - (\mathbb{E}[X])^2$$

## Esempi di Calcolo della Varianza

### Esempio 1: Lancio di un dado equo
Consideriamo una variabile aleatoria $X$ che rappresenta il risultato del lancio di un dado equo a 6 facce.

1. Calcoliamo $\mathbb{E}[X]$:
   $$\mathbb{E}[X] = \sum_{i=1}^6 i \cdot \frac{1}{6} = \frac{21}{6} = 3.5$$

2. Calcoliamo $\mathbb{E}[X^2]$:
   $$\mathbb{E}[X^2] = \sum_{i=1}^6 i^2 \cdot \frac{1}{6} = \frac{91}{6} \approx 15.167$$

3. Quindi la varianza è:
   $$\text{Var}[X] = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \frac{91}{6} - (\frac{21}{6})^2 = \frac{35}{12} \approx 2.917$$

### Esempio 2: Variabile aleatoria di Bernoulli
Sia $X$ una variabile aleatoria di Bernoulli con parametro $p$ (successo con probabilità $p$, insuccesso con probabilità $1-p$).

1. Sappiamo che:
   $$\mathbb{E}[X] = p$$

2. Poiché $X$ può assumere solo i valori 0 e 1:
   $$\mathbb{E}[X^2] = 0^2 \cdot (1-p) + 1^2 \cdot p = p$$

3. Quindi:
   $$\text{Var}[X] = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = p - p^2 = p(1-p)$$

## Note Importanti

1. La varianza ha unità di misura quadratiche rispetto alla variabile originale
2. Per questo motivo, spesso si usa la deviazione standard $ \text{Std}[X] = \sqrt{\text{Var}[X]}$
3. La formula $\mathbb{E}[X^2] - (\mathbb{E}[X])^2$ è spesso più pratica per i calcoli
4. La varianza è particolarmente utile nello studio di somme di variabili aleatorie indipendenti

# La Distribuzione Binomiale

Immaginiamo di voler lanciare una moneta 3 volte e di voler contare il numero di teste.

* Comprendiamo il Processo
  - Ogni lancio è indipendente
  - La probabilità di testa è $p$ (successo)
  - La probabilità di croce è $1-p$ (insuccesso)
  - Parliamo dell'evento "ottenere $k$ successi in $n$ lanci"

* Iniziamo con un Caso Concreto: Supponiamo di volere esattamente 2 teste in 3 lanci.Come può succedere?
    1. Testa, Testa, Croce (TTC)
    2. Testa, Croce, Testa (TCT)
    3. Croce, Testa, Testa (CTT)

Notiamo che:
    - Ogni sequenza ha la stessa probabilità
    - Stiamo parlando di esiti diversi.
    - Tutte queste sequenze fanno riferimento all'evento "otteniamo 2 teste in 3 lanci".

Calcoliamo la Probabilità di una Sequenza Specifica
Prendiamo TTC come esempio:
  - Primo lancio (T): probabilità $p$
  - Secondo lancio (T): probabilità $p$
  - Terzo lancio (C): probabilità $(1-p)$

Probabilità totale della sequenza = $p \cdot p \cdot (1-p) = p^2(1-p)$. Non è difficile rendersi conto che questa probabilità vale per tutte e tre le sequenze di sopra, quindi la probabilità dell'evento "otteniamo 2 teste in 3 lanci" è: $3 \cdot p^2(1-p)$

> Riconosciamo il Pattern Il numero 3 non è casuale. È il numero di modi di scegliere 2 posizioni da 3 possibili! Oppure, può anche essere inteso come il numero di disposizioni di 3 elementi dove ci sono due classi di elementi indistinguibili (T e C) e dove ci sono 2 elementi di tipo T e 1 di tipo C. Vi ricordate il caso degli anagrammi, sì?! Insomma, il numero che cerchiamo lo abbiamo usando:
$$\frac{3!}{2!1!}=\binom{3}{2} = 3$$

* Generalizziamo. Per $k$ successi in $n$ prove:
  1. Ogni sequenza specifica ha probabilità $p^k(1-p)^{n-k}$
  2. Il numero di possibili sequenze è $\binom{n}{k}$

Quindi la probabilità totale è:

$$P(X = k) = \binom{n}{k}p^k(1-p)^{n-k}$$

* Per chiudere il cerchio, calcoliamo la probabilità di ottenere esattamente 2 teste in 3 lanci con una moneta equa ($p = 1/2$):

1. Calcoliamo $\binom{3}{2} = 3$
2. $p^2 = (1/2)^2 = 1/4$
3. $(1-p)^{3-2} = (1/2)^1 = 1/2$

Quindi:
$$P(X = 2) = 3 \cdot \frac{1}{4} \cdot \frac{1}{2} = \frac{3}{8}$$

Si noti che, data l'assunzione di equiprobabilità dei risultati, il risultato coincide con il calcolo del semplice rapporto tra casi favorevoli e casi totali!

## Proprietà della distribuzione Binomiale:
1. **Media**: $\mathbb{E}[X] = np$
2. **Varianza**: $\text{Var}[X] = np(1-p)$

Note:
1. La somma delle probabilità per tutti i possibili valori di $k$ è 1:
   $$\sum_{k=0}^n \binom{n}{k}p^k(1-p)^{n-k} = 1$$
   (Questo è il teorema binomiale!)

2. I casi estremi hanno senso:
   - $P(X = 0) = (1-p)^n$ (tutti insuccessi)
   - $P(X = n) = p^n$ (tutti successi)

La distribuzione binomiale è un modello fondamentale perché emerge ogni volta che contiamo successi in prove indipendenti.

## La Variabile Aleatoria di Bernoulli

La variabile aleatoria di Bernoulli è il caso più semplice di variabile aleatoria discreta e descrive un esperimento con due soli possibili esiti: successo (1) o insuccesso (0).

### Definizione Formale
Una variabile aleatoria $X$ si dice di Bernoulli se:
- $X$ può assumere solo i valori 0 e 1
- $P(X = 1) = p$, dove $p \in [0,1]$ è il parametro della distribuzione
- $P(X = 0) = 1 - p$

La funzione di probabilità può essere scritta in forma compatta come:
$$P(X = k) = p^k(1-p)^{1-k}, \quad k \in \{0,1\}$$

### Caratteristiche
- **Valore Atteso**: $\mathbb{E}[X] = p$
- **Varianza**: $\text{Var}(X) = p(1-p)$

## Generalizzazione alla Variabile Aleatoria Binomiale

La variabile aleatoria binomiale può essere vista come una naturale estensione della variabile di Bernoulli: rappresenta la somma di $n$ prove di Bernoulli indipendenti e identicamente distribuite (i.i.d.).

### Il Collegamento tra Bernoulli e Binomiale

Se consideriamo una sequenza di $n$ esperimenti di Bernoulli indipendenti, ciascuno con probabilità di successo $p$, e definiamo:
- $X_1, X_2, ..., X_n$ come le singole prove di Bernoulli
- $Y = \sum_{i=1}^n X_i$ come il numero totale di successi

Allora $Y$ segue una distribuzione binomiale con parametri $n$ e $p$, che scriviamo come $Y \sim B(n,p)$.

La funzione di probabilità della binomiale è:
$$P(Y = k) = \binom{n}{k}p^k(1-p)^{n-k}, \quad k \in \{0,1,...,n\}$$

### Dimostrazione della Generalizzazione

1. Ogni singola prova $X_i$ è una Bernoulli$(p)$
2. La somma $Y$ rappresenta il conteggio dei successi in $n$ prove
3. Per le proprietà di indipendenza:
   - $\mathbb{E}[Y] = \mathbb{E}[\sum_{i=1}^n X_i] = \sum_{i=1}^n \mathbb{E}[X_i]$
   - $\text{Var}(Y) = \text{Var}(\sum_{i=1}^n X_i) = \sum_{i=1}^n \text{Var}(X_i)$

Questo ci porta alle formule per la binomiale:
- $\mathbb{E}[Y] = np$
- $\text{Var}(Y) = np(1-p)$

che sono chiaramente una generalizzazione delle formule della Bernoulli per $n=1$.

### Osservazioni Importanti

1. La v.a. di Bernoulli è un caso particolare della binomiale quando $n=1$
2. La binomiale eredita e amplifica le proprietà statistiche della Bernoulli
3. Le proprietà di indipendenza sono fondamentali:
   $$P(X_i = 1 | X_j = k) = P(X_i = 1) = p, \quad \forall i \neq j$$

Questa generalizzazione è fondamentale per:
- L'analisi di sequenze di prove ripetute
- Lo studio di fenomeni discreti in vari ambiti applicativi
- La modellazione di processi con esiti binari multipli
