Ricordiamo che qualsiasi distribuzione discreta è concentrata su un numero finito o numerabile di valori isolati. Al contrario, le variabili continue possono assumere qualsiasi valore in un intervallo, $(a, b)$, $(a, +\infty)$, $(-\infty, +\infty)$, ecc. Vari tempi come il tempo di servizio, il tempo di installazione, il tempo di download, il tempo di guasto, e anche misure fisiche come peso, altezza, distanza, velocità, temperatura e velocità di connessione sono esempi di **variabili aleatorie continue**.

## 4.1  Densità di probabilità

![Figure 4.1](figs/4.1.png)

Per tutte le variabili continue, la funzione di massa di probabilità (PMF) è sempre uguale a zero:  

$ P(x) = 0 $  

per ogni $ x $.  

Di conseguenza, la funzione di massa di probabilità non fornisce alcuna informazione riguardo a una **variabile aleatoria**. In alternativa, possiamo utilizzare la **funzione di distribuzione cumulativa** (CDF) $ F(x) $. Nel caso continuo, essa è data da  
$$ F(x) = P(X \leq x) = P(X < x)  $$

Si noti che queste due espressioni per $ F(x) $ differiscono per il termine $ P(X = x) = P(x) = 0 $, che è nullo.  

Nei casi sia continuo che discreto, la CFD $ F(x) $ è una funzione non decrescente che assume valori compresi tra 0 e 1. Come visto nel Capitolo 3, nel caso discreto il grafico di $ F(x) $ presenta salti di ampiezza pari a $ P(x) $. Per le distribuzioni continue, poiché $ P(x) = 0 $, non vi sono salti. Pertanto, in questo caso la funzione di distribuzione cumulativa è una funzione continua.

Assumiamo, inoltre, che $ F(x) $ abbia una derivata. Questo è il caso per tutte le distribuzioni continue comunemente usate, ma in generale non è garantito dalla continuità e monotonicità (la famosa funzione di Cantor è un controesempio).

> **DEFINIZIONE 4.1**  La **funzione di densità di probabilità** (PDF, densità) è la derivata della CDF,  $ f(x) = F'(x) $. La distribuzione si dice continua se possiede una densità.  
Allora, $ F(x) $ è un'antiderivata di una densità. Per il Teorema Fondamentale del Calcolo, l'integrale di una densità da $ a $ a $ b $ è uguale alla differenza delle antiderivate, ovvero  
$$
 \int_{a}^{b} f(x) \, dx = F(b) - F(a) = P(a < X < b) ,
$$  
dove notiamo nuovamente che la probabilità a destra è uguale anche a $ P(a \leq X < b) $, $ P(a < X \leq b) $ e $ P(a \leq X \leq b) $.

____________________
**Funzione di densità di probabilità**  

$$
    f(x) = F'(x)
$$  
$$
 P(a < X < b) = \int_{a}^{b} f(x) \, dx
$$  
___________________

Pertanto, le probabilità possono essere calcolate integrando una densità sugli insiemi dati. Inoltre, l'integrale  
$$
 \int_{a}^{b} f(x) \, dx
$$  
è uguale all'area sotto la curva di densità tra i punti $ a $ e $ b $. Geometricamente, quindi, le probabilità sono rappresentate da aree (Figura 4.1). Sostituendo $ a = -\infty $ e $ b = +\infty $, otteniamo  
$$
 \int_{-\infty}^{b} f(x) \, dx = P(-\infty < X < b) = F(b) \quad \text{e} \quad \int_{-\infty}^{+\infty} f(x) \, dx = P(-\infty < X < +\infty) = 1 .
$$  
Ovvero, l'area totale sotto la curva di densità è uguale a 1.  

Guardando la Figura 4.1, possiamo vedere perché $ P(X = x) = 0 $ per tutte le **variabili aleatorie** continue. Questo perché  
$$
 P(X = x) = P(x \leq X \leq x) = \int_{x}^{x} f(t) \, dt = 0 .
$$

geometricamente, è l'area sotto la curva di densità, dove due lati della regione collassano in uno solo.


> **Esempio 4.1.** La durata, in anni, di un componente elettronico è una **variabile aleatoria** continua con la densità  
$$
    f(x) = 
    \begin{cases}
    \dfrac{k}{x^3} & \text{per } x \geq 1 \\
    0 & \text{per } x < 1 .
    \end{cases}
$$  
Trovare $ k $, disegnare il grafico della cdf $ F(x) $ e calcolare la probabilità che la durata superi i 5 anni.

**Soluzione.** Si trova $ k $ dalla condizione  
$$
 \int_{-\infty}^{+\infty} f(x) \, dx = 1 :
$$  
$$
 \int_{1}^{+\infty} \frac{k}{x^3} \, dx = k \int_{1}^{+\infty} x^{-3} \, dx = k \left[ -\frac{1}{2x^2} \right]_{1}^{+\infty} = \frac{k}{2} = 1 .
$$  
Quindi, $ k = 2 $. Integrando la densità, si ottiene la cdf,  
$$
 F(x) = \int_{-\infty}^{x} f(y) \, dy = \int_{1}^{x} \frac{2}{y^3} \, dy = 2 \left[ -\frac{1}{2y^2} \right]_{1}^{x} = 1 - \frac{1}{x^2}
$$  
per $ x > 1 $. Il suo grafico è mostrato nella Figura 4.2.

![Figure 4.2](figs/4.2.png)


Successivamente, calcoliamo la probabilità che la durata superi i 5 anni,  
$$
 P(X > 5) = 1 - F(5) = 1 - \left(1 - \frac{1}{5^2}\right) = \frac{1}{25} = 0.04 .
$$  

Possiamo anche ottenere questa probabilità integrando la densità,  
$$
 P(X > 5) = \int_{5}^{+\infty} f(x) \, dx = \int_{5}^{+\infty} \frac{2}{x^3} \, dx = 2 \left[ -\frac{1}{2x^2} \right]_{5}^{+\infty} = \frac{1}{25} = 0.04 .
$$  
♦

$$  
Questa volta, se l'intera regione sotto la curva di densità viene ritagliata da un pezzo di legno, essa si bilancerà in un punto con coordinata $ E(X) $, come mostrato nella Figura 4.3.

Dimostriamo che  
$$
 \operatorname{Var}(X) = E(X - \mu)^2 = \int (x - \mu)^2 f(x) \, dx = E(X^2) - \mu^2.
$$

Partiamo dalla definizione:  
$$
 \operatorname{Var}(X) = E(X - \mu)^2 = \int_{-\infty}^{+\infty} (x - \mu)^2 f(x) \, dx ,
$$  
dove $ \mu = E(X) $.  

Sviluppiamo il quadrato:  
$$
 (x - \mu)^2 = x^2 - 2\mu x + \mu^2 .
$$  
Quindi,  
$$
 \int (x - \mu)^2 f(x) \, dx = \int (x^2 - 2\mu x + \mu^2) f(x) \, dx = \int x^2 f(x) \, dx - 2\mu \int x f(x) \, dx + \mu^2 \int f(x) \, dx .
$$  

Ora valutiamo i tre integrali:  
- $ \int x f(x) \, dx = E(X) = \mu $,  
- $ \int f(x) \, dx = 1 $ (poiché $ f $ è una densità),  
- $ \int x^2 f(x) \, dx = E(X^2) $.  

Sostituendo, otteniamo:  
$$
 \operatorname{Var}(X) = E(X^2) - 2\mu \cdot \mu + \mu^2 \cdot 1 = E(X^2) - 2\mu^2 + \mu^2 = E(X^2) - \mu^2 .
$$

> **Esempio 4.2.** Una **variabile aleatoria** $ X $ dell'Esempio 4.1 ha densità  
$ f(x) = 2x^{-3} $ per $ x \geq 1 $.  

Il suo valore atteso è  
$$
 \mu = E(X) = \int x f(x) \, dx = \int_{1}^{\infty} 2x^{-2} \, dx = \left[ -2x^{-1} \right]_{1}^{\infty} = 2 .
$$


Calcolando la sua varianza, ci imbattiamo in una “sorpresa”,  
$$
 \sigma^2 = \operatorname{Var}(X) = \int x^2 f(x) \, dx - \mu^2 = \int_{1}^{\infty} 2x^{-1} \, dx - 4 = 2 \ln x \Big|_{1}^{\infty} - 4 = +\infty .
$$  
Questa variabile non ha una varianza finita! (Si veda anche la discussione sul paradosso di San Pietroburgo a p. 62.)

## 4.2 Distribuzioni di Probabilità Continue
Come nel caso discreto, una varietà di fenomeni può essere descritta da un numero relativamente ridotto di famiglie di distribuzioni continue. Qui discuteremo le famiglie Uniforme, Esponenziale e Normale.


### 4.2.1 Distribuzione Uniforme

La distribuzione Uniforme gioca un ruolo unico nella modellazione stocastica. Una variabile aleatoria con qualsiasi distribuzione pensabile può essere generata a partire da una variabile aleatoria Uniforme. Molti linguaggi di programmazione e software sono dotati di un generatore di numeri pseudo-casuali che produce variabili aleatorie Uniformi. Gli utenti possono poi convertirle in variabili con distribuzioni desiderate tramite manipolazioni deterministiche e utilizzarle per la simulazione al computer di vari eventi e processi.

Inoltre, la distribuzione Uniforme è utilizzata in qualsiasi situazione in cui un valore viene scelto "a caso" da un dato intervallo; cioè, senza alcuna preferenza per valori inferiori, superiori o medi. Ad esempio, le posizioni degli errori in un programma, i compleanni durante un anno e molte variabili aleatorie continue modulo 1, modulo 0.1, 0.01, ecc., sono distribuite uniformemente sui loro corrispondenti intervalli.

Per dare uguale preferenza a tutti i valori, la distribuzione Uniforme ha una densità costante (Figura 4.4). Sull'intervallo $ (a, b) $, la sua densità è uguale a

$$
f(x) = \frac{1}{b - a}, \quad a < x < b,
$$

perché l'area rettangolare sotto il grafico della densità deve essere uguale a 1.

Per la stessa ragione, $ |b - a| $ deve essere un numero finito. Non esiste una distribuzione Uniforme sull'intera retta reale. In altre parole, se ti viene chiesto di scegliere un numero casuale da $ (-\infty, +\infty) $, non puoi farlo in modo uniforme.

![Figure 4.4](figs/4.4.png)

**La proprietà Uniforme**

Per ogni $ h > 0 $ e $ t \in [a, b - h] $, la probabilità

$$
P(t < X < t + h) = \int_t^{t+h} \frac{1}{b-a} \, dx = \frac{h}{b-a}
$$

non dipende da $ t $. Questa è la proprietà Uniforme: la probabilità è determinata soltanto dalla lunghezza dell'intervallo, e non dalla sua posizione.



**Esempio 4.3.** Nella Figura 4.4, i rettangoli A e B hanno la stessa area, il che mostra che

$$
P(s < X < s + h) = P(t < X < t + h).
$$
♦

**Esempio 4.4.** Se un volo previsto all'arrivo alle 17:00 in realtà arriva in un istante distribuito uniformemente tra le 16:50 e le 17:10, allora è altrettanto probabile che arrivi prima delle 17:00 o dopo le 17:00, altrettanto probabile prima delle 16:55 o dopo le 17:05, ecc.
♦


**Distribuzione Uniforme Standard**
  
La distribuzione Uniforme con $ a = 0 $ e $ b = 1 $ è chiamata distribuzione Uniforme Standard.  
La densità della distribuzione Uniforme Standard è $ f(x) = 1 $ per $ 0 < x < 1 $. La maggior parte dei generatori di numeri casuali restituisce una variabile aleatoria Uniforme Standard.

Tutte le distribuzioni Uniformi sono legate tra loro nel modo seguente. Se $ X $ è una variabile aleatoria Uniforme su $ (a, b) $, allora

$$
Y = \frac{X - a}{b - a}
$$

è Uniforme Standard. Analogamente, se $ Y $ è Uniforme Standard, allora

$$
X = a + (b - a)Y
$$

è Uniforme su $ (a, b) $. Si verifichi che $ X \in (a, b) $ se e solo se $ Y \in (0, 1) $.

Numerose altre famiglie di distribuzioni possiedono un membro “standard”. In generale, una semplice trasformazione consente di convertire una variabile aleatoria standard in una non standard, e viceversa.


**Valore atteso e varianza**  
Per una variabile aleatoria $ Y $ Uniforme Standard,

$$
E(Y) = \int y f(y) \, dy = \int_0^1 y \cdot 1 \, dy = \frac{1}{2}
$$

e

$$
\text{Var}(Y) = E(Y^2) - E^2(Y) = \int_0^1 y^2 \, dy - \left( \frac{1}{2} \right)^2 = \frac{1}{3} - \frac{1}{4} = \frac{1}{12}.
$$

Ora consideriamo il caso generale. Sia $ X = a + (b - a)Y $, dove $ Y $ è Uniforme Standard, per cui $ X $ ha distribuzione Uniforme$ (a, b) $.  
Utilizzando le proprietà del valore atteso e della varianza date in (3.5) e (3.7),

$$
E(X) = E\{a + (b - a)Y\} = a + (b - a)E(Y) = a + (b - a) \cdot \frac{1}{2} = \frac{a + b}{2}
$$

e

$$
\text{Var}(X) = \text{Var}\{a + (b - a)Y\} = (b - a)^2 \text{Var}(Y) = (b - a)^2 \cdot \frac{1}{12} = \frac{(b - a)^2}{12}.
$$





Il valore atteso corrisponde esattamente al punto medio dell'intervallo $ [a, b] $. Il fatto che non vi sia preferenza tra parti sinistre o destre è in accordo con la proprietà Uniforme e con il significato fisico di $ E(X) $ come centro di gravità.

| Distribuzione Uniforme $ (a, b) $ |  |
|-------------------------------|---|
| Valori possibili               | $ a < x < b $ |
| Funzione di densità $ f(x) $   | $ \frac{1}{b - a} $ |
| Valore atteso $ E(X) $         | $ \frac{a + b}{2} $ |
| Varianza $ \text{Var}(X) $     | $ \frac{(b - a)^2}{12} $ |

### 4.2.2. Variabile Aleatoria Esponenziale

Immaginiamo un tipo di evento di Poisson (esempio: chiamate a un call center, decadimenti di un atomo radioattivo). Supponiamo che questi eventi siano quindi caratterizzati da un unico parametro: il **tasso** $ \lambda $.

-   $ \lambda $: Il **numero medio di eventi che si verificano per unità di tempo** (es. 2 chiamate al minuto, 0.1 guasti all'ora).
-   Ora creiamo una variabile aleatoria $ X_t $, che **conta il numero di eventi in un intervallo di lunghezza $ t $**, e rendiamoci conto che essa ha una **distribuzione di Poisson** con parametro $ \lambda t $.

    La sua funzione di massa di probabilità (PMF) è:
    $$
    P(X_t = k) = e^{-\lambda t} \frac{(\lambda t)^k}{k!}
    $$
    dove $ \lambda t $ rappresenta il numero atteso di eventi in un intervallo di lunghezza $ t $.


Ora cambiamo prospettiva. Invece di chiederci "quanti eventi avvengono in un tempo fisso $ t $?", ci chiediamo: "quanto tempo $ T $ devo aspettare per osservare il *prossimo* evento?".

-   $ T $: Una **variabile aleatoria continua** che rappresenta il **tempo tra due eventi consecutivi** (o il tempo fino al primo evento).

> L'obiettivo è trovare la funzione di densità di probabilità di $ T $ (PDF) e, per farlo, colleghiamo  la nuova variabile $ T $ (tempo) alla variabile che già conosciamo, $ X_t $. 

Nello specifico, consideriamo la probabilità $ P(T > t) $, ovvero la probabilità di dover aspettare *più tempo* rispetto a quello in  $ t $ per il prossimo evento. L'osservazione cruciale è che l'evento $ \{T > t\} $ ("il primo evento avviene dopo l'istante $ t $") è logicamente **equivalente** all'evento $ \{X_t = 0\} $ ("nessun evento si verifica nell'intervallo $ [0, t] $"). E quindi:
$$
P(T > t) = P(X_t = 0)
$$

Sappiamo dalla distribuzione di Poisson di $X_t$ che:

$$
P(X_t = 0) = e^{-\lambda t} \frac{(\lambda t)^0}{0!} = e^{-\lambda t}
$$

E, di conseguenza:
$$
P(T > t) = e^{-\lambda t}
$$


Questo non ci dice ancora nulla sulla funzione di densità di $T$, ma sì sulla funzione di ripartizione (CDF). Infatti, la CDF di qualsiasi variabile aleatoria $ T $, $ F_T(t) $, è definita come la probabilità che $ T $ assuma un valore minore o uguale a $ t $:
$$
F_T(t) = P(T \le t)
$$

Per le proprietà della probabilità sappiamo che:
$$
F_T(t) = P(T \le t) = 1 - P(T > t)
$$

Ora, usando il fatto che $P(T>t) = P(X=0)$, possiamo ricavare:
 
$$
F_T(t) = P(T < t) = 1 - P(T > t) = 1 - P(X=0)
$$
 
da cui:
 
$$
F_T(t) = P(T < t) = 1 - P(X=0) = 1 - e^{-\lambda t}
$$


Abbiamo appena trovato la CDF di una v.a. continua $T$ che indica il _tempo tra due eventi di Poisson la cui frequenza è $\lambda\,t$ per intervalli di lunghezza $t$_. Facciamo subito un'osservazione:

Noi sappiamo trasformare distribuzioni di Poisson che contano uno stesso tipo di evento quando l'intervallo di riferimento cambia di grandezza: si tratta solo di applicare il corrispondente fattore multiplicativo alla frequenza. Per esempio, se una v.a. di Poisson che conta il numero di eventi ogni 5 minuti ha frequenza $\lambda=2$, sappiamo che possiamo creare un'altra v.a. che conti il numero dello stesso tipo di evento ogni $5\times3=15$ minuti usando una frequenza $\tilde{\lambda}=2\times3=6$. 

Non è quindi difficile rendersi conto che, se $T$ indica il tempo tra due eventi di Poisson la cui frequenza è $\lambda\,t$ per intervalli di lunghezza $t$, essa indica anche il tempo tra due eventi di Poisson la cui frequenza è $\lambda$ per intervalli di lunghezza unitaria. Infatti stiamo parlando degli stessi eventi. E quando prima abbiamo detto che $ X_t \sim \text{Poisson}(\lambda t) $, abbiamo implicitamente affermato che il tasso unitario sottostante è $ \lambda $. Pertanto, la variabile "tempo di attesa" $ T $ eredita naturalmente questo stesso tasso $ \lambda $.

La tabella seguente chiarisce la relazione tra l'intervallo di osservazione e il parametro di Poisson:

| Lunghezza Intervallo | Numero Atteso di Eventi | Parametro di Poisson |
| :------------------- | :----------------------: | :------------------: |
| $ 1 $              |         $ \lambda $         |      $ \lambda $       |
| $ t $              |        $ \lambda t $        |      $ \lambda t $     |


Abbiamo così dimostrato che quando il  **numero di eventi** di Poisson in un intervallo di lunghezza unitaria segue una CDF **Poisson($ \lambda $)**, il **tempo di attesa** tra eventi consecutivi segue una CDF  $ F_T(t) = 1 - e^{-\lambda t} $.

Ma quale è la PDF di $T$? Partiamo dalla  CDF che abbiamo precedentemente derivato:

$$
F_T(t) = P(T \leq t) = 
\begin{cases} 
1 - e^{-\lambda t}, & t \geq 0 \\
0, & t < 0 
\end{cases}
$$
(perché il tempo tra due eventi non può essere negativo). 

Ora noi sappiamo che, per una variabile aleatoria continua, la PDF $ f_T(t) $, è definita come la derivata della CDF rispetto alla sua variabile:
$$
f_T(t) = \frac{d}{dt} F_T(t)
$$

Applichiamo ora questa definizione alla nostra CDF:

-   Per $ t < 0 $:
    $$
    F_T(t) = 0 \quad \Rightarrow \quad f_T(t) = \frac{d}{dt}(0) = 0
    $$

-   Per $ t \geq 0 $:
    $$
    \begin{aligned}
    f_T(t) &= \frac{d}{dt} \left( 1 - e^{-\lambda t} \right) \\
    &= \frac{d}{dt}(1) - \frac{d}{dt}\left(e^{-\lambda t}\right) \\
    &= 0 - \left( -\lambda e^{-\lambda t} \right) \\
    &= \lambda e^{-\lambda t}
    \end{aligned}
    $$



Combinando i due casi, otteniamo la PDF cercata:

$$
f_T(t) = 
\begin{cases} 
\lambda e^{-\lambda t}, & t \geq 0 \\
0, & t < 0 
\end{cases}
$$

Una buona verifica è controllare che l'integrale della PDF su tutto l'asse reale sia uguale a 1, come deve essere per qualsiasi densità di probabilità valida.

$$
\begin{aligned}
\int_{-\infty}^{\infty} f_T(t)  dt &= \int_{0}^{\infty} \lambda e^{-\lambda t}  dt \\
&= \lambda \left[ -\frac{1}{\lambda} e^{-\lambda t} \right]_{0}^{\infty} \\
&= \left[ -e^{-\lambda t} \right]_{0}^{\infty} \\
&= (0) - (-1) = 1
\end{aligned}
$$

La verifica è positiva, confermando che la PDF derivata è corretta.

Grafichiamo queste funzioni e notiamo due cose:

````python
import numpy as np
import matplotlib.pyplot as plt

def plot_interarrival_poisson_time_cdf_and_pdf(lambda_param):
    # Generazione dei dati
    x = np.linspace(0, 5, 100)
    pdf = lambda_param * np.exp(-lambda_param * x)
    cdf = 1 - np.exp(-lambda_param * x)

    # Creazione del grafico
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    

    ax[0].plot(x, pdf, label='PDF', color='blue')
    ax[0].set_title('Funzione di Densità di Probabilità (PDF)')
    ax[0].set_xlabel('x')
    ax[0].set_ylabel('f(x)')
    ax[0].axhline(0, color='black', lw=0.5, ls='--')
    ax[0].axvline(0, color='black', lw=0.5, ls='--')
    ax[0].legend()

    ax[1].plot(x, cdf, label='CDF', color='red')
    ax[1].set_title('Funzione di Distribuzione (CDF)')
    ax[1].set_xlabel('x')
    ax[1].set_ylabel('F(x)')
    ax[1].axhline(0, color='black', lw=0.5, ls='--')
    ax[1].axvline(0, color='black', lw=0.5, ls='--')
    ax[1].legend()

    fig.suptitle(f'Tempo di inter-arrivo tra eventi di Poisson con parametro (λ={lambda_param})')
    plt.legend()
    plt.grid()
    plt.show()
````

-   La **PDF** $ f_T(t) $ è massima all'istante $ t=0 $ (con valore $ \lambda $) e decade "**_esponenzialmente_**", riflettendo il fatto che in un processo di Poisson è più probabile che un evento accada a breve termine piuttosto che dopo un lungo periodo.


> **Ebbene sì, la variabile aleatoria continua che indica il tempo tra due eventi di Poisson si chiama variabile aleatoria esponenziale.** Ed è proprio quella $T$ di cui stiamo parlando. 

La variabile aleatoria esponenziale è una delle distribuzioni di probabilità più importanti in statistica e teoria della probabilità, particolarmente utile nel campo dell'informatica e dell'ingegneria. Questa distribuzione è comunemente utilizzata per modellare il tempo tra eventi in un processo di Poisson, dove gli eventi si verificano in modo indipendente e a una media costante.

La distribuzione esponenziale è stata introdotta nel contesto dei processi stocastici e trova applicazione in vari campi, tra cui:

- **Teoria delle code**: per modellare il tempo di attesa in sistemi di servizio.
- **Reliability engineering**: per analizzare il tempo fino al guasto di un sistema o di un componente.
- **Processi di Poisson**: per descrivere il tempo tra eventi in processi che seguono una distribuzione Poisson.

In sintesi: La variabile aleatoria esponenziale è viene utlizzata per modellare il tempo tra eventi che si verificano con un tasso costante nel tempo.


**Funzione di densità di probabilità (PDF)**:
  
  La funzione di densità di probabilità della variabile aleatoria esponenziale è data da:

  $$
  f(x; \lambda) = \lambda e^{-\lambda x} \quad \text{per } x \geq 0 \tag{4.1}
  $$ 

  dove $ \lambda > 0 $ è l'unico parametro della distribuzione, noto come tasso di occorrenza o frequenza.

**Funzione cumulativa di distribuzione (CDF)**:
  
  La funzione cumulativa è data da:

  $$
  F(x; \lambda) = \int_0^x f(x)\,dx =  \int_0^x \lambda e^{-\lambda t} dt =  1 - e^{-\lambda x} \quad \text{per } x \geq 0 \tag{4.2}
  $$



**Valore atteso**:
  
  Il valore atteso $ \mathbb{E}[X] $ della variabile aleatoria esponenziale è:

  $$
    \mathbb{E}[X] = \int x\,f(x)dx = \int_0^\infty t\,\lambda e^{-\lambda t} dt = \frac{1}{\lambda} \tag{4.3}
  $$

**Dimostrazione**
Infatti, per sviluppare questo integrale:

$$
\mathbb{E}[X] = \int_0^\infty t \cdot \lambda e^{-\lambda t}  dt
$$

Possiamo usare la formula per l'integrazione per parti è $ \int u  dv = uv - \int v  du $.
Poniamo:
- $ u = t $ → $ du = dt $
- $ dv = \lambda e^{-\lambda t} dt $ → $ v = -e^{-\lambda t} $

Applicando la formula avremo:
$$
\begin{aligned}
\mathbb{E}[X] &= \int_0^\infty t \cdot \lambda e^{-\lambda t}  dt \\
&= \left[ t \cdot (-e^{-\lambda t}) \right]_{0}^{\infty} - \int_0^\infty (-e^{-\lambda t})  dt \\
&= \left[ -t e^{-\lambda t} \right]_{0}^{\infty} + \int_0^\infty e^{-\lambda t}  dt
\end{aligned}
$$

- Valutiamo il termine $ \left[ -t e^{-\lambda t} \right]_{0}^{\infty} $:
    - Per $ t \to \infty $: $ \lim_{t \to \infty} (-t e^{-\lambda t}) = 0 $ (l'esponenziale decresce più velocemente di qualsiasi potenza di $ t $).
    - Per $ t = 0 $: $ -(0)e^{0} = 0 $.
    - Quindi, $ \left[ -t e^{-\lambda t} \right]_{0}^{\infty} = 0 $.

- Calcoliamo l'integrale rimanente:
$$
\int_0^\infty e^{-\lambda t}  dt = \left[ -\frac{1}{\lambda} e^{-\lambda t} \right]_{0}^{\infty} = (0) - \left( -\frac{1}{\lambda} \right) = \frac{1}{\lambda}
$$


Combinando i risultati:
$$
\mathbb{E}[X] = 0 + \frac{1}{\lambda} = \frac{1}{\lambda}
$$


$$ \square $$



**Varianza**
  
  $$
  Var(X) = \int x^2\,f(x)dx - \mathbb{E}[X]^2 = \int_0^\infty t^2\,\lambda e^{-\lambda t} dt - \left(\frac{1}{\lambda}\right)^2 = \frac{1}{\lambda^2} - \frac{1}{\lambda^2} = \frac{1}{\lambda^2} \tag{4.4}
  $$

**Dimostrazione**

Infatti, la varianza è data da $ Var(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 $. Conosciamo già $ \mathbb{E}[X] = \frac{1}{\lambda} $, quindi dobbiamo calcolare $ \mathbb{E}[X^2] $:

$$
\mathbb{E}[X^2] = \int_0^\infty t^2 \cdot \lambda e^{-\lambda t}  dt
$$

Identifichiamo nuovamente le parti per l'integrazione per parti. Poniamo:
- $ u = t^2 $ → $ du = 2t  dt $
- $ dv = \lambda e^{-\lambda t} dt $ → $ v = -e^{-\lambda t} $

Applichiamo la formula:
$$
\begin{aligned}
\mathbb{E}[X^2] &= \int_0^\infty t^2 \cdot \lambda e^{-\lambda t}  dt \\
&= \left[ t^2 \cdot (-e^{-\lambda t}) \right]_{0}^{\infty} - \int_0^\infty (-e^{-\lambda t}) \cdot 2t  dt \\
&= \left[ -t^2 e^{-\lambda t} \right]_{0}^{\infty} + 2 \int_0^\infty t e^{-\lambda t}  dt
\end{aligned}
$$

- Valutiamo il termine $ \left[ -t^2 e^{-\lambda t} \right]_{0}^{\infty} $:
    - Per $ t \to \infty $: $ \lim_{t \to \infty} (-t^2 e^{-\lambda t}) = 0 $.
    - Per $ t = 0 $: $ -(0)^2 e^{0} = 0 $.
    - Quindi, $ \left[ -t^2 e^{-\lambda t} \right]_{0}^{\infty} = 0 $.

- Notiamo che l'integrale rimanente $ \int_0^\infty t e^{-\lambda t}  dt $ è esattamente il valore atteso che abbiamo appena calcolato, diviso per $ \lambda $. Più precisamente:
$$
\int_0^\infty t e^{-\lambda t}  dt = \frac{\mathbb{E}[X]}{\lambda} = \frac{1/\lambda}{\lambda} = \frac{1}{\lambda^2}
$$
(In alternativa, si può risolvere con un'ulteriore integrazione per parti, arrivando allo stesso risultato).

Quindi l'espressione per $ \mathbb{E}[X^2] $ diventa:
$$
\mathbb{E}[X^2] = 0 + 2 \cdot \frac{1}{\lambda^2} = \frac{2}{\lambda^2}
$$

Infine, applicando la definizione alternativa della varianza otteniamo

$$
\begin{aligned}
Var(X) &= \mathbb{E}[X^2] - (\mathbb{E}[X])^2 \\
&= \frac{2}{\lambda^2} - \left( \frac{1}{\lambda} \right)^2 \\
&= \frac{2}{\lambda^2} - \frac{1}{\lambda^2} \\
&= \frac{1}{\lambda^2}
\end{aligned}
$$

come volevasi dimostrare.

$$ \square $$



La quantità $ \lambda $ è un parametro della distribuzione Esponenziale, e il suo significato risulta chiaro dalla relazione $ \mathbb{E}(X) = 1/\lambda $. Se $ X $ rappresenta un tempo, misurato in minuti, allora $ \lambda $ è una frequenza, misurata in min $^{-1}$.

Ad esempio, se gli eventi o arrivi si verificano in media ogni mezzo minuto, allora $ \mathbb{E}(X) = 0.5 $ e $ \lambda = 2 $, il che indica una frequenza (tasso di arrivo) di 2 arrivi al minuto. Si noti che questo $ \lambda $ ha lo stesso significato del parametro della distribuzione di Poisson presentato nella Sezione 3.4.5.


**Esempio 4.5.** Dei lavori vengono inviati a una stampante con un tasso medio di 3 lavori all’ora.  
(a) Qual è il tempo atteso tra un lavoro e il successivo?  
(b) Qual è la probabilità che il prossimo lavoro venga inviato entro 5 minuti?

**Soluzione.** Gli arrivi dei lavori rappresentano eventi rari; pertanto, il tempo $ T $ tra di essi è distribuito in modo Esponenziale con parametro $ \lambda = 3 \ \text{h}^{-1} $ (lavori all’ora).  

(a) $ \mathbb{E}(T) = 1/\lambda = 1/3 $ ore, ovvero 20 minuti tra un lavoro e l’altro;  

(b) Convertiamo nell’unità di misura comune: 5 minuti = $ 1/12 $ ore. Allora,

$$
P(T < 1/12 \ \text{h}) = F(1/12) = 1 - e^{-\lambda (1/12)} = 1 - e^{-3 \cdot (1/12)} = 1 - e^{-1/4} = 0.2212.
$$
♦

**Assenza di memoria**: La distribuzione esponenziale ha la proprietà unica di "assenza di memoria", che significa che la probabilità che un evento si verifichi in un certo intervallo di tempo non dipende da quanto tempo è già passato. Si dice che “le variabili Esponenziali perdono la memoria”. Cosa significa?

Supponiamo che una variabile Esponenziale $ T $ rappresenti un tempo di attesa. La proprietà di assenza di memoria (_memoryless property_) significa che il fatto di aver già atteso per $ t $ minuti viene “dimenticato”, e non influenza il tempo di attesa futuro. Indipendentemente dall’evento $ T > t $, quando il tempo di attesa totale supera $ t $, il tempo di attesa residuo ha ancora distribuzione Esponenziale con lo stesso parametro.

Se questo suona un po' strano, non preoccuparti, guarda cosa significa in termini quantitativi, e tutto diventa ovvio:

$$
P(T > t + x \mid T > t) = P(T > x)
\quad \text{per } t, x > 0.
\quad \text{(4.6)}
$$

In questa formula, $ t $ rappresenta la parte già trascorsa del tempo di attesa, mentre $ x $ è il tempo aggiuntivo, quello residuo.

**Dimostrazione:** La (4.2) non è scontata. Per dimostrarla formalmente, partiamo da, $ P(T > x) = e^{-\lambda x} $. Inoltre, usando la formula (2.7) per la probabilità condizionata,

$$
P(T > t + x \mid T > t) = \frac{P\{T > t + x \cap T > t\}}{P(T > t)}
$$

È qui che le cose diventano interessanti, perché, stando ai fatti, $P\{T > t + x \cap T > t\} = P\{T > t + x \}$. (Se il tempo di attesa è maggiore di $T + x$ è anche maggiore di $x$. O, in altre parole, l'evento $T > t$ è un sottoinsieme proprio dell'evento $T > t + x$). Quindi possiamo dire che: 

$$
P(T > t + x \mid T > t) = \frac{P\{T > t + x \cap T > t\}}{P(T > t)} = \frac{P(T > t + x)}{P(T > t)} = \frac{e^{-\lambda(t + x)}}{e^{-\lambda t}} = e^{-\lambda x}.
$$
$\Box$

Questa proprietà è unica della distribuzione Esponenziale. Nessun'altra variabile aleatoria continua $ X \in (0, \infty) $ è priva di memoria. Tra le variabili discrete, una proprietà analoga appartiene alla distribuzione Geometrica (vedi il capitolo bonus 3.3.3) In un certo senso, la distribuzione Geometrica è l’analogo discreto della distribuzione Esponenziale. 

| Distribuzione Esponenziale |  |
|----------------------------|---|
| Parametro $ \lambda $ | frequenza, numero di eventi per unità di tempo |
| Funzione di densità $ f(x) $ | $ \lambda e^{-\lambda x},\ x > 0 $ |
| Valore atteso $ E(X) $ | $ \frac{1}{\lambda} $ |
| Varianza $ \text{Var}(X) $ | $ \frac{1}{\lambda^2} $ |

La variabile aleatoria esponenziale è ampiamente utilizzata in vari ambiti dell'informatica, grazie alle sue proprietà uniche e alla capacità di modellare fenomeni temporali. Di seguito sono elencate alcune delle applicazioni pratiche più significative.

**Teoria delle Code** La distribuzione esponenziale è fondamentale nella teoria delle code, dove viene utilizzata per modellare il tempo di attesa in sistemi di servizio. Ad esempio:
- **Sistemi di prenotazione**: Il tempo che un cliente attende prima di essere servito può essere modellato come una variabile esponenziale.
- **Server e richieste**: In un sistema informatico, il tempo tra le richieste dei clienti a un server può seguire una distribuzione esponenziale, facilitando l'analisi delle prestazioni del sistema[1][4].

**Reliability Engineering** Nella ingegneria della affidabilità, la distribuzione esponenziale viene utilizzata per descrivere il tempo fino al guasto di componenti e sistemi. Le applicazioni includono:
- **Analisi dei guasti**: La durata di vita di dispositivi elettronici o meccanici è spesso modellata con una distribuzione esponenziale, specialmente quando i guasti sono rari e non dipendono dall'età del dispositivo[2][5].
- **Manutenzione predittiva**: Utilizzando modelli esponenziali, le aziende possono prevedere quando è probabile che si verifichino guasti e pianificare la manutenzione di conseguenza.

**Simulazione di Processi Stocastici** La distribuzione esponenziale è utile nella simulazione di processi stocastici, in particolare:
- **Generazione di numeri casuali**: In molti algoritmi di simulazione Monte Carlo, i tempi tra eventi casuali vengono generati utilizzando variabili aleatorie esponenziali per rappresentare il tempo tra eventi discreti[2].
- **Modellazione di traffico**: Nella simulazione del traffico urbano o delle reti di comunicazione, il tempo tra arrivi di veicoli o pacchetti dati può essere rappresentato da una distribuzione esponenziale[4][5].
- **Modelli predittivi**: Utilizzata per modellare il tempo fino a un evento significativo (ad esempio, l'acquisto di un prodotto da parte di un cliente) in modelli predittivi.

**In conclusione** La variabile aleatoria esponenziale è uno strumento versatile nell'informatica, con applicazioni che spaziano dalla teoria delle code alla reliability engineering fino alla simulazione e all'analisi dei dati. La sua capacità di modellare eventi temporali rende questa distribuzione particolarmente utile in contesti dove il tempo gioca un ruolo cruciale.
