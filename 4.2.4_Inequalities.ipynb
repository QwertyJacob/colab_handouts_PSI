### 4.3.1 Disuguaglianze di Markov e di Čebyšëv
__________________

Finora abbiamo studiato distribuzioni specifiche (come la Binomiale, la Uniforme o la Normale) in cui, conoscendo i parametri, potevamo calcolare con precisione la probabilità di ogni evento. 

Ma cosa succede se non conosciamo la forma della distribuzione? Se sappiamo solo che una variabile ha una certa media $\mu$ e magari una varianza $\sigma^2$, cosa possiamo dire sulle sue probabilità?

Le **disuguaglianze probabilistiche** forniscono dei "limiti" (bound) universali. Anche se non sono precise come un calcolo sulla Gaussiana, hanno il potere di essere vere per **qualsiasi** variabile aleatoria, indipendentemente dalla sua forma.

**Osservazione Storica:**  
Pafnutij L'vovič Čebyšëv e il suo allievo Andrej Andreevič Markov facevano parte della celebre "scuola russa" di San Pietroburgo nel XIX secolo. Il loro lavoro ha gettato le basi per la legge dei grandi numeri, dimostrando che il caso, su larga scala, segue regole deterministiche sorprendenti.

#### Disuguaglianza di Markov

La disuguaglianza di Markov è la più semplice e richiede una sola condizione: che la variabile aleatoria $X$ non sia mai negativa.

> **TEOREMA 4.4 (Disuguaglianza di Markov)**   Sia $X$ una variabile aleatoria non negativa ($X \ge 0$). 
>
> Per ogni costante $\epsilon > 0$ si avrà:
> $$P(X \ge \epsilon) \le \frac{\mathbb{E}[X]}{\epsilon}$$

**L'intuizione:** Questa disuguaglianza ci dice che se la media di una popolazione è bassa, non possono esserci troppi individui con valori estremamente alti. Ad esempio, se il reddito medio è 1000€, è matematicamente impossibile che più del 10% della popolazione guadagni più di 10.000€.

**Dimostrazione (Logica dell'indicatrice)**

Per capire perché è vera, definiamo una variabile "indicatrice" $\mathbb{I}$:
$$
\mathbb{I} = \begin{cases} 1 & \text{se } X \ge \epsilon \\ 0 & \text{se } X < \epsilon \end{cases}
$$
Notiamo che vale sempre la relazione $X \ge \epsilon \cdot \mathbb{I}$. Infatti:
- Se $X < \epsilon$, allora $\mathbb{I}=0$ e la relazione diventa $X \ge 0$ (vero).
- Se $X \ge \epsilon$, allora $\mathbb{I}=1$ e la relazione diventa $X \ge \epsilon$ (vero).

Applicando la proprietà di **monotonia del valore atteso**:

$$\mathbb{E}[X] \ge \mathbb{E}[\epsilon \cdot \mathbb{I}] = \epsilon \cdot \mathbb{E}[\mathbb{I}]$$

Poiché $\mathbb{E}[\mathbb{I}]$ è semplicemente la probabilità $P(X \ge \epsilon)$, otteniamo:

$$\mathbb{E}[X] \ge \epsilon \cdot P(X \ge \epsilon) \implies P(X \ge \epsilon) \le \frac{\mathbb{E}[X]}{\epsilon}$$

#### Disuguaglianza di Čebyšëv

Mentre Markov usa solo la media, Čebyšëv sfrutta anche la **varianza** per dirci quanto una variabile è "concentrata" attorno alla sua media. A differenza di Markov, questa vale per **qualsiasi** distribuzione (anche quelle che assumono valori negativi).

> **TEOREMA 4.5 (Disuguaglianza di Čebyšëv)**  
> Sia $X$ una variabile aleatoria con media $\mu$ e varianza $\sigma^2$. Per ogni $\lambda > 0$:
>
> $$P(|X - \mu| \ge \lambda\sigma) \le \frac{1}{\lambda^2}$$

In altre parole, la probabilità che una variabile si allontani dalla media per più di $\lambda$ deviazioni standard è al massimo $1/\lambda^2$.

**Esempio di interpretazione:** Per qualsiasi variabile aleatoria con con media $\mu$ e varianza $\sigma^2$: 
- Per $\lambda = 2$: Al massimo il $25\%$ dei valori è fuori dall'intervallo $[\mu - 2\sigma, \mu + 2\sigma]$.
- Per $\lambda = 3$: Al massimo l' $11.1\%$ dei valori è fuori dall'intervallo $[\mu - 3\sigma, \mu + 3\sigma]$.


**Dimostrazione:** Una delle bellezze della matematica è la sua modularità. Una volta dimostrata la disuguaglianza di Markov, dimostrare quella di Čebyšëv diventa un esercizio di **sostituzione strategica**.

**Teorema:** Sia $X$ una variabile aleatoria con media $\mu$ e varianza $\sigma^2$. Allora per ogni $k > 0$:
$$P(|X - \mu| \ge k) \le \frac{\sigma^2}{k^2}$$

*Nota: Nella formula standard usiamo spesso $k = \lambda\sigma$ per esprimere la distanza in "numero di deviazioni standard".*

**Dimostrazione:**

1. Consideriamo l'evento di cui vogliamo stimare la probabilità: $|X - \mu| \ge k$.

2. Poiché entrambi i membri della disuguaglianza sono positivi, possiamo elevarli al quadrato senza cambiare il senso della disuguaglianza:
   $$(X - \mu)^2 \ge k^2$$

3. Definiamo una nuova variabile aleatoria $Y = (X - \mu)^2$. Notiamo che $Y$ è, per costruzione, una **variabile non negativa**. 

4. Possiamo quindi applicare la **Disuguaglianza di Markov** a $Y$, usando come soglia il valore $\epsilon = k^2$:
   $$P(Y \ge k^2) \le \frac{\mathbb{E}[Y]}{k^2}$$

5. Sostituiamo ora a ritroso:
   - $P(Y \ge k^2)$ è identico a $P(|X - \mu| \ge k)$.
   - $\mathbb{E}[Y]$ è $\mathbb{E}[(X - \mu)^2]$, che per definizione è la **Varianza** di $X$ ($\sigma^2$).
6
. Otteniamo così il risultato finale:
   $$P(|X - \mu| \ge k) \le \frac{\sigma^2}{k^2}$$

$$\square$$


**La "regola empirica" vs Čebyšëv**
Ricordi la distribuzione Normale? Per una Gaussiana, sappiamo che dentro $\pm 2\sigma$ cade circa il **95%** dei dati. Čebyšëv è più conservativo: ci garantisce che per *qualsiasi* distribuzione (anche la più strana possibile) dentro $\pm 2\sigma$ cadrà **almeno il 75%** dei dati.

````python
import numpy as np
import matplotlib.pyplot as plt

# Visualizzazione del Bound di Chebyshev
lambdas = np.linspace(1.1, 5, 100)
chebyshev_bound = 1 / lambdas**2

plt.figure(figsize=(10, 5))
plt.plot(lambdas, chebyshev_bound, 'r-', label='Limite massimo (Čebyšëv)')
plt.fill_between(lambdas, 0, chebyshev_bound, color='red', alpha=0.1)
plt.title('Disuguaglianza di Čebyšëv: Probabilità nelle "code"')
plt.xlabel('Distanza dalla media (in deviazioni standard λ)')
plt.ylabel('P(|X - μ| ≥ λσ)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
````

> **Esempio: Controllo Qualità** Le ante di legno di un lotto industriale hanno una lunghezza media $\mu = 205$ cm e una varianza $\sigma^2 = 8$ cm². Un'anta è difettosa se la sua lunghezza è minore di 197 cm o superiore a 213 cm. Il lotto viene rifiutato se contiene più del 15% di ante difettose. Senza conoscere la distribuzione, il lotto verrà rigettato?

**Soluzione:**
1. **Dati:** $\mu = 205$, $\sigma = \sqrt{8} \approx 2.83$.
2. **Intervallo di tolleranza:** L'intervallo è $[197, 213]$. Notiamo che è simmetrico rispetto alla media: $205 \pm 8$.
3. **Calcolo di $\lambda$:** Vogliamo sapere a quante deviazioni standard corrisponde la soglia di 8 cm.
   $$\lambda\sigma = 8 \implies \lambda = \frac{8}{2.83} \approx 2.82$$
4. **Applicazione di Čebyšëv:**
   $$P(|X - 205| \ge 8) \le \frac{1}{2.82^2} \approx \frac{1}{7.95} \approx 0.125$$
5. **Conclusione:** La probabilità di avere un'anta difettosa è **al massimo del 12.5%**. Poiché il limite per il rigetto del lotto è il 15%, sulla base delle informazioni a disposizione, il lotto **non sarà rigettato** (poiché $12.5\% < 15\%$).

Nata: In questo esercizio, abbiamo fatto un passaggio logico che merita attenzione. Calcoliamo la probabilità che **un'anta specifica** sia difettosa ($P \le 12.5\%$) e da lì concludiamo che **il lotto** (composto da migliaia di ante) probabilmente non verrà rigettato. 

Perché possiamo farlo? Quali proprietà ci permettono questo "salto" dal micro al macro?

- **Identica Distribuzione e Indipendenza (i.i.d.)** Assumiamo che ogni anta sia prodotta dallo stesso processo industriale. Matematicamente, trattiamo le caratteristiche di ogni anta come variabili aleatorie $X_1, X_2, \dots, X_n$ che sono **Indipendenti e Identicamente Distribuite**.
- **Proprietà della proporzione, o Legge dei Grandi Numeri** Il legame tra la probabilità del singolo evento e la percentuale di difetti nel lotto è regolato dalla **Linearità del Valore Atteso**. Se definiamo una variabile $D_i$ che vale $1$ se l'anta $i$-esima è difettosa e $0$ se è integra, il numero totale di difetti nel lotto è $K = \sum D_i$.
Il valore atteso della *proporzione* di difetti ($K/n$) sarà quindi:
$$\mathbb{E}\left[\frac{K}{n}\right] = \frac{1}{n} \sum \mathbb{E}[D_i] = \frac{1}{n} \cdot n \cdot P(\text{difetto}) = P(\text{difetto})$$

Quindi, se la probabilità che un'anta sia difettosa è $\le 12.5\%$, ci aspettiamo che **in media** il $12.5\%$ delle ante nel lotto sia difettoso. Vedremo più sulla legge dei grandi numeri nel capitolo 4.4.

**NB:** È fondamentale chiarire un punto: Čebyšëv non ci garantisce che *un singolo lotto specifico* non supererà mai il 15% di scarti. Ci dice che, date le informazioni (media e varianza), la **proporzione attesa** di ante difettose è inferiore alla soglia di tolleranza. In altre parole, quando diciamo che il lotto non sarà rigettato, intendiamo che sulla base delle informazioni a disposizione, il processo è statisticamente sicuro.
Poiché la stima peggiore della probabilità di difetto $(12.5\%)$ è inferiore alla soglia critica ($15\%$), il valore atteso dei difetti nel lotto rientra nei limiti di accettabilità. Solo una fluttuazione statistica molto sfortunata (e tanto più rara quanto più è grande il lotto) potrebbe portare al rigetto.
